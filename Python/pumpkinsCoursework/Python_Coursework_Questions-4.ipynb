### Question 7
Analyze the life expectancy dataset (which can be loaded by running the cell below).

- Calculate the number of unique countries in the dataset.

- Calculate the country and year with the highest life expectancy.

- Calculate the year with the highest total expenditure.

- Provide your code using pandas.


````python
import seaborn as sns
life_expectancy = sns.load_dataset('healthexp')
````

````python
import pandas as pd

# part 1
num_countries = life_expectancy['Country'].nunique()
print(num_countries)

# part 2
highest_life = life_expectancy.loc[life_expectancy['Life_Expectancy'].idxmax()]
print(highest_life)

# part 3
highest_exp_year = life_expectancy.groupby('Year')['Spending_USD'].sum().idxmax()
print(highest_exp_year)
````

### Question 10
Write a function that will convert carat to grams and find the heaviest diamond.

1 carat = 0.2 grams

````python
def carat_to_grams(carat):
    return carat * 0.2
diamonds["weight_grams"] = diamonds["carat"].apply(carat_to_grams)
heaviest_diamond = diamonds.loc[diamonds["weight_grams"].idxmax()]
print(heaviest_diamond)
````
